```python
from hydra import compose, initialize
with initialize(version_base=None, config_path="./object_detection/DETR"):
    cfg = compose(config_name="detr")
```

```python
from PIL import Image
import requests
img = Image.open(requests.get(cfg.data.sample_url, stream=True).raw)
img
```

```python
from object_detection.inference import inference
from object_detection.inference import rescale_bbox
from object_detection.plot import plot_results
feature_extractor, model, encoding, outputs = inference(img, cfg)
probas, keep, bboxes_scaled = rescale_bbox(
    img, outputs, feature_extractor, cfg)
plot_results(img, model, probas[keep], bboxes_scaled)
```

```python
from object_detection.inference import get_attention_weights
from object_detection.plot import visualize_attention
conv_features, dec_attn_weights = get_attention_weights(model, encoding)
visualize_attention(img, model, probas, keep, conv_features,
                    bboxes_scaled, dec_attn_weights)
```
